```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
crime_data_2023 = {
    "Murder": 5,
    "Violent Crime": 380,
    "Property Crime": 1900,
    "Robbery": 100,
    "Burglary": 250,
    "Motor Vehicle Theft": 300
}

plt.figure(figsize=(8,5))
plt.bar(crime_data_2023.keys(), crime_data_2023.values(), color="skyblue")
plt.title("Crime Rates in 2023 (per 100k people)")
plt.ylabel("Rate per 100,000")
plt.xticks(rotation=0)
plt.show()
```

```python
crime_changes = {
    "Murder": -13,
    "Violent Crime": -6,
    "Property Crime": -4,
    "Robbery": -15,
    "Burglary": -12,
    "Motor Vehicle Theft": -10
}

plt.figure(figsize=(8,5))
plt.barh(list(crime_changes.keys()), list(crime_changes.values()), color="lightcoral")
plt.title("% Change in Crime Rates (2022 â†’ 2023)")
plt.xlabel("% Change")
plt.axvline(x=0, color="black", linewidth=0.8)
plt.show()
```

```python
companies = ["Company A", "Company B", "Company C"]
products = ["Product 1", "Product 2", "Product 3"]

data = {
    "Company A": [10, 15, 20],
    "Company B": [12, 18, 25],
    "Company C": [8, 14, 22]
}

x = np.arange(len(products))
width = 0.25

fig, ax = plt.subplots(figsize=(8,5))

ax.bar(x - width, data["Company A"], width, label="Company A")
ax.bar(x, data["Company B"], width, label="Company B")
ax.bar(x + width, data["Company C"], width, label="Company C")

ax.set_title("Grouped Bar Chart (Fake Company Data)")
ax.set_xticks(x)
ax.set_xticklabels(products)
ax.set_ylabel("Values")
ax.legend()

plt.show()
```
