```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x = np.linspace(-10, 10, 400)

# Define the function f(x) = 1/x^2 + 1.
f_x = 1 / (x**2 + 1)

# Plot with tight axis
plt.figure()
plt.plot(f_x, x)
plt.title('Plot with plt.axis(\'tight\')')
plt.xlabel('f(x)')
plt.ylabel('x')
plt.grid(True)
plt.axis('tight')
plt.gca().invert_yaxis() # Invert the y-axis
plt.gca().invert_xaxis() # Invert the x-axis


# Plot with equal axis
plt.figure()
plt.plot(f_x, x)
plt.title('Plot with plt.axis(\'equal\')')
plt.xlabel('f(x)')
plt.ylabel('x')
plt.grid(True)
plt.axis('equal')
plt.gca().invert_yaxis()
plt.gca().invert_xaxis()


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Define the range for x, only for x >= 0
x = np.linspace(0, 10, 400)
# Define the functions
f_x = np.sqrt(x)
g_x = np.power(x, 1/3)

plt.plot(x, f_x, label='f(x) = sqrt(x)')
plt.plot(x, g_x, label='g(x) = x^(1/3)')

plt.title('Plot of f(x) = sqrt(x) and g(x) = x^(1/3)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 50 random points for x and y
np.random.seed(42) # for reproducibility
x = np.random.rand(50)
y = np.random.rand(50)

# Create a scatter plot using plt.plot() with the 'o' marke
plt.figure()
plt.plot(x, y, 'o')
plt.title('Scatter plot using plt.plot()')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Create the same scatter plot using plt.scatter()
plt.figure()
plt.scatter(x, y)
plt.title('Scatter plot using plt.scatter()')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data for the function f(x) = x*sin(x) over the range [0, 4*pi]
x = np.linspace(0, 4 * np.pi, 100)
y = x * np.sin(x)

# Create a scatter plot with a line connecting the points
plt.plot(x, y, marker='o', linestyle='-',
         markersize=8, linewidth=2,
         markerfacecolor='lightblue', markeredgecolor='darkblue',
         markeredgewidth=1.5)

plt.title('Plot of f(x) = x*sin(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x = np.linspace(-np.pi, 2 * np.pi, 400)

# Define the function f(x) = x*cos(x)
f_x = x * np.cos(x)

# Define x values at every pi/4 interval within the range
x_scatter = np.arange(-np.pi, 2 * np.pi + np.pi/4, np.pi/4)
f_x_scatter = x_scatter * np.cos(x_scatter)

# Plot the function with a blue dashed line
plt.plot(x, f_x, color='blue', linestyle='--', label='f(x) = x*cos(x)')
# Add scatter points with red circular markers
plt.scatter(x_scatter, f_x_scatter, color='red', marker='o', zorder=5, label='Points at pi/4 intervals')

plt.title('Plot of f(x) = x*cos(x) with different styles')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
